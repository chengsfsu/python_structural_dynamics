## Dynamic magnification factor

\begin{equation}
DMF = \frac{1}{\sqrt{(1-\beta)^2 + (2 \xi \beta)^2 }}
\end{equation}

- Let's assume a frequency ratio betweeb 0 and 2.
- Initially asume 5% of damping ratio: $\xi$ = 0.05%.


```python
# Dependencies
import numpy as np
import matplotlib.pyplot as plt

# Generate a vector of frequency ratios
max_beta = 2                                # Maximun value of beta
delta_beta = 0.01                           # Step size in range of beta values
nPoints = int(max_beta / delta_beta)        # Number of data points -> int
beta = np.linspace(0, max_beta, nPoints) 

# Damping ratio
xi = 0.05

# Vector of dynamic magnification factor
DMF = 1 / np.sqrt((1 - beta**2)**2 + (2 * xi * beta)**2)

# Plotting
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 2, 1.5])
axes.plot(beta, DMF, "r-", label="5% Damping")
axes.set_title("Dynamic Magnification Factor")
axes.set_xlabel("Frequency ratio $\\beta$")
axes.set_xlim([0, 2])
axes.set_ylabel("DMF")
axes.legend()
axes.grid()
fig.show()
```

###Observations

- When $\omega$ is close to $\omega_n$, the DMF peaks.
- DMF peaks is approximately 10 for 5% of damping.

### Considering a range of damping ratios

```python
# Range of damping ratio
Xi = np.array([0.05, 0.1, 0.2, 0.25, 0.5, 0.75])

# Creating plot base
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 2, 1.5])
axes.set_title("Dynamic Magnification Factor")
axes.set_xlabel("Frequency ratio $\\beta$")
axes.set_xlim([0, 2])
axes.set_ylabel("DMF")
axes.grid()

# Plotting diferent damping ratio
for xi in Xi:
    DMF = 1 / np.sqrt((1 - beta**2)**2 + (2 * xi * beta)**2)
    axes.plot(beta, DMF, label=f"{int(xi*100)}% Damping")
    axes.legend()

# Showing result
fig.show()
```
